```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
area= np.array([350,430,455,557,640,700,770,860,900])
price= np.array([11,22,41,38,57,74,70,79,88])
```

```python
area
```

```python
plt.figure(figsize=(10,3))
plt.xlabel("Area in Squarefeet")
plt.ylabel("Price in lakhs")
plt.title("Area vs Price plot")
plt.scatter(area, price)
plt.show()
```

```python
area_mean=area.mean()
price_mean=price.mean()
area_mean, price_mean
```

```python
coef=np.sum((area-area_mean)*(price-price_mean))/np.sum(np.square(area-area_mean))
```

```python
coef
```

```python
intercept=price_mean-area_mean*coef
intercept
```

```python
#y=mx+c
y_pred=coef*area+intercept
y_pred
```

```python
plt.figure(figsize=(10,3))
plt.xlabel("Area in Squarefeet")
plt.ylabel("Price in lakhs")
plt.title("Area vs Price plot")
plt.scatter(area, price)
plt.plot(area, price)
plt.scatter(area, y_pred)
plt.plot(area, y_pred)
plt.show()
```

```python
MSE=np.sum(np.square(price-y_pred))/len(price)
MSE
```

```python
MAE=np.sum(np.abs(price-y_pred))/len(price)
MAE
```

```python
SSE= np.sum(np.square(price-y_pred))
SSE
```
