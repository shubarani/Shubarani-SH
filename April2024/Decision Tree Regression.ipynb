```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.datasets import make_regression
```

```python
X1,y1=make_regression(n_samples=1000,n_features=5,noise=0.3)
```

```python
X1
```

```python
y1
```

```python
X_train1,X_test1,y_train1,y_test1=train_test_split(X1,y1,test_size=0.25,random_state=43)
```

```python
model=DecisionTreeRegressor()
```

```python
model.fit(X_train1,y_train1)
```

```python
y_pred1=model.predict(X_test1)
```

```python
mse=mean_squared_error(y_test1,y_pred1)
```

```python
mse
```

```python
r2_score(y_test1,y_pred1)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
(X,y)=make_regression(n_samples=1000,n_features=5,noise=0.3)
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=43)
```

```python
model=DecisionTreeRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
r2_score(y_test,y_pred)
```
