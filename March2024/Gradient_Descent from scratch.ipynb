```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area= np.array([350,430,455,557,640,700,770,860,900])
price= np.array([11,22,41,38,57,74,70,79,88])
```

```python
plt.figure(figsize=(10,3))
plt.xlabel("Area in Squarefeet")
plt.ylabel("Price in lakhs")
plt.title("Area vs Price plot")
plt.scatter(area, price)
plt.show()
```

```python
b=14
X=area
y=price
m=0.13282388809365295
```

```python
y_pred = m*X+b
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred)
plt.show()
```

```python
b_slope= -2*np.sum(y-m*X-b)
```

```python
b_slope
```

```python
lr=0.1
```

```python
b=b-lr*b_slope
```

```python
b
```

```python
y_pred2=m*X+b
```

```python
plt.scatter(X,y)
plt.plot(X,y_pred,label="b 1")
plt.plot(X,y_pred2, label="b 2")
plt.legend()
plt.show()
```
