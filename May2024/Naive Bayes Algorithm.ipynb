```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv(r"C:\Users\Shuba\Downloads\iris_prac.csv")
```

```python
df.head()
```

```python
X=df.drop("Target",axis=1)
```

```python
X
```

```python
y=df["Target"]
```

```python
y
```

```python
X_tr,X_te,y_tr,y_te=train_test_split(X,y,train_size=0.75,random_state=43)
```

```python
model=GaussianNB()
```

```python
model.fit(X_tr,y_tr)
```

```python
model.score(X_tr,y_tr)
```

```python
y_pr=model.predict(X_te)
```

```python
accuracy_score(y_te,y_pr)
```

```python
confusion_matrix(y_te,y_pr)
```

```python
print(classification_report(y_te,y_pr))
```

```python
sns.heatmap(confusion_matrix(y_te,y_pr),annot=True)
```
